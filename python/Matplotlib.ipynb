```python
import matplotlib.pyplot as plt
```

```python
# xlabel()
# ylabel()
# title

x = ['Jan','Feb','March','April','May','June']
y =[ 10 , 20, 40, 30,45,10]
plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Temp')
plt.title('Month Vs Temp')
plt.show()
```

```python
# line color
# marker
# marker size
x = ['Jan','Feb','March','April','May','June']
y =[ 10 , 20, 40, 30,45,10]
plt.plot(x,y , color='r',marker ='o',markersize=10)
plt.xlabel('Month')
plt.ylabel('Temp')
plt.title('Month Vs Temp')
plt.show()
```

```python
#  let suppose i need to compare 
# the data for 2 years 
x = ['Jan','Feb','March','April','May','June']
y_2020 =[ 10 , 20, 40, 30,45,10]
y_2021 =[ 40 , 30, 40, 70,25,60]
plt.plot(x,y_2020 , color='r',marker ='o',markersize=10)
plt.plot(x,y_2021 , color='b',marker ='o',markersize=10)
plt.xlabel('Month')
plt.ylabel('Temp')
plt.title('Month Vs Temp')
plt.show()
```

```python
# plt.legend()
x = ['Jan','Feb','March','April','May','June']
y_2020 =[ 10 , 20, 40, 30,45,10]
y_2021 =[ 40 , 30, 40, 70,25,60]
plt.plot(x,y_2020 , color='r',marker ='o',markersize=10 , label = '2020')
plt.plot(x,y_2021 , color='b',marker ='o',markersize=10, label = '2021')
plt.xlabel('Month')
plt.ylabel('Temp')
plt.title('Month Vs Temp')
plt.legend()
plt.show()
```

```python
# read data set 
# and analyse Glucose Vs age
# diabetic : 1 and Non diabetic :0
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.head()
```

```python
# filter the value for 1
df_d = df[df.Outcome==1]
df_d.head()
```

```python
# filter the value for 0
df_nd = df[df.Outcome==0]
df_nd.head()
```

```python
# 
plt.figure(figsize=(10,5))

# subplot #1
plt.subplot(1,2,1)
# Diabetic
plt.scatter(df_d.Age,df_d.Glucose,c='r',s=5)
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('diabetic')
# subplot #1
plt.subplot(1,2,2)
plt.scatter(df_nd.Age,df_nd.Glucose,c='b',s=5)
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('non-diabetic')

plt.show()
```
